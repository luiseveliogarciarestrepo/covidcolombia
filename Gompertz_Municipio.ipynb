```python
municipio = 'Manizales'
```

```python
gompertz.columns
```

```python
filtro = gompertz['Ciudad de ubicación'] == municipio
gompertz_mun = gompertz.loc[filtro,:]
gompertz_mun
```

```python
gompertz_mun=gompertz_mun.append(df_fechas).sort_values(by=['Fecha de notificación'], ignore_index=True)
gompertz_mun
```

gompertz_mun.dropna(how='any', subset=['Ciudad de ubicación',
       'Departamento o Distrito '], inplace=True)
gompertz_mun

```python
longitud = gompertz_mun['ID de caso'].size
longitud
```

```python
gompertz_mun.columns
```

```python
for i in range(0, longitud):
    if gompertz_mun.loc[i, 'ID de caso'] == 0:
        pass
    else:
        gompertz_mun.loc[i, 'ID de caso'] = 1
```

```python
gompertz_mun
```

```python
gompertz_mun['ID de caso'].sum()
```

```python
gompertz_mun['Acumulados']= 0
gompertz_mun['Día número']=0
gompertz_mun['Penetración real']=0
gompertz_mun
```

```python
gompertz_mun.loc[0, 'Acumulados']= gompertz_mun.loc[0, 'ID de caso']
for i in range(1, longitud):
    gompertz_mun.loc[i,'Acumulados'] = gompertz_mun.loc[i, 'ID de caso'] + gompertz_mun.loc[i-1, 'Acumulados']
```

```python
gompertz_mun
```

```python
gompertz_mun.columns
```

```python
gompertz_mun = gompertz_mun.groupby('Fecha de notificación')[['ID de caso', 'Ciudad de ubicación',
       'Departamento o Distrito ', 'Acumulados', 'Día número',
       'Penetración real']].max()
```

```python
gompertz_mun.reset_index(inplace=True)
```

```python
longitud = gompertz_mun['ID de caso'].size
longitud
```

```python
poblacion = poblacionGral.loc[poblacionGral['Municipio']==municipio,'2020'].sum()
poblacion
```

```python
gompertz_mun.loc[0, 'Día número'] = 1
for i in range(1, longitud):
    gompertz_mun.loc[i,'Día número'] = i+1
    gompertz_mun.loc[i,'Penetración real'] = (gompertz_mun.loc[i, 'Acumulados']/ poblacion)*100
    
```

```python
gompertz_mun.drop(columns=['Ciudad de ubicación', 'Departamento o Distrito '], inplace=True)
```

```python
gompertz_mun
```

```python
gompertz_mun.drop(columns=['ID de caso', 'Acumulados'], inplace=True)
```
